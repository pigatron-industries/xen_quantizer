### Dissonance Curve Comparison

Comparison of 31 EDO with 12 EDO against the dissonance curve for harmonic sounds.

```python
from lib.tuning import Tuning
from lib.scale import Scale, Chord
import lib.tuningplot as tuningplot
%matplotlib inline

tuning_12edo = Tuning.createEqualDivisionTuning("12edo", 12, 1, ['0 C','1 C#','2 D','3 D#','4 E','5 F','6 F#','7 G','8 G#','9 A','10 A#','11 B'])
tuning_31edo = Tuning.createEqualDivisionTuning("31edo", 31, 1)

plot = tuningplot.TuningPlot('h', 14)
plot.plotRatios([(1, 1), (2, 1), (3, 2), (4, 3), (5, 3), (5, 4), (6, 5), (7, 4), (7, 5), (8,5), (9, 8), (9, 5)])
plot.plotDissonance([1, 2, 3, 4, 5, 6, 7, 8, 9], 5.5)
plot.plotTuning(tuning_12edo)
plot.plotTuning(tuning_31edo)

plot.plot()
```

### Orwell [9]

```python
tuning_31edo = Tuning.createEqualDivisionTuning("31edo", 31, 1)
scale_orwell = Scale.fromSteps("Orwell [9]", tuning_31edo, [4, 3, 4, 3, 4, 3, 4, 3, 3])
chord_23 = Chord(scale_orwell, [0, 2, 5])
chord_32 = Chord(scale_orwell, [0, 3, 5])

# Scale
plot = tuningplot.TuningPolarPlot(4, 4, 5, 19)
plot.plotRatios([(1, 1), (3, 2), (5, 3), (5, 4), (6, 5), (7, 4), (8, 5), (9, 8), (11, 8)])
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "Orwell", True)
plot.plot()

# Chord 0-Subminor
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_23, 0, "Subminor")
plot.plotRatios([(1, 1), (3, 2), (6, 5)])
plot.plot()
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_32, 0, "Supermajor")
plot.plotRatios([(1, 1), (3, 2), (5, 4)])
plot.plot()

plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_23, 1, "Diminished\nSubminor")
plot.plotRatios([(1, 1), (3, 2), (6, 5)], 4/31)
plot.plot()
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_32, 1, "Diminished\nMajor")
plot.plotRatios([(1, 1), (3, 2), (5, 4)], 4/31)
plot.plot()

plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_23, 2, "Subminor")
plot.plotRatios([(1, 1), (3, 2), (6, 5)], 7/31)
plot.plot()
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_32, 2, "Supermajor")
plot.plotRatios([(1, 1), (3, 2), (5, 4)], 7/31)
plot.plot()

plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_23, 3, "Diminished\nSubminor")
plot.plotRatios([(1, 1), (3, 2), (6, 5)], 11/31)
plot.plot()
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_32, 3, "Diminished\nMajor")
plot.plotRatios([(1, 1), (3, 2), (5, 4)], 11/31)
plot.plot()

plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_23, 4, "Diminished\nSubminor")
plot.plotRatios([(1, 1), (3, 2), (6, 5)], 14/31)
plot.plot()
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_32, 4, "Diminished\nSupermajor")
plot.plotRatios([(1, 1), (3, 2), (5, 4)], 14/31)
plot.plot()

plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_23, 5, "Diminished\nSubminor")
plot.plotRatios([(1, 1), (3, 2), (6, 5)], 18/31)
plot.plot()
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_32, 5, "Diminished\nMajors")
plot.plotRatios([(1, 1), (3, 2), (5, 4)], 18/31)
plot.plot()

plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_23, 6, "Diminished\nSubminor")
plot.plotRatios([(1, 1), (3, 2), (6, 5)], 21/31)
plot.plot()
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_32, 6, "Diminished\nSupermajor")
plot.plotRatios([(1, 1), (3, 2), (5, 4)], 21/31)
plot.plot()

plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_23, 7, "")
plot.plotRatios([(1, 1), (3, 2), (6, 5), (9, 8)], 25/31)
plot.plot()
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_32, 7, "")
plot.plotRatios([(1, 1), (3, 2), (5, 4)], 25/31)
plot.plot()

plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_23, 8, "Diminished\nSubminor")
plot.plotRatios([(1, 1), (3, 2), (6, 5)], 28/31)
plot.plot()
plot.plotTuning(tuning_31edo)
plot.plotScale(scale_orwell, 0, 2, "", True)
plot.plotChord(chord_32, 8, "Diminished\nMajor")
plot.plotRatios([(1, 1), (3, 2), (5, 4)], 28/31)
plot.plot()
```

## Isomorphic Keyboard Layouts

### Bosanquet-Wilson

```python
from lib.tuning import Tuning
from lib.HexPlot import HexPlot

tuning_31edo = Tuning.createEqualDivisionTuning("31edo", 31, 1)
root = tuning_31edo.getNote(3, 0)

plot = HexPlot(tuning_31edo, width=10, height=14)
plot.highlightNote("ROOT", root, "red")
# # plot.highlightNote("m-2ND", 6, 1, "violet")
# # plot.highlightNote("M-2ND", 7, 1, "blueviolet")
plot.highlightNote("mm-3RD", root.add(7), "orange")
plot.highlightNote("m-3RD", root.add(8), "orange")
plot.highlightNote("M-3RD", root.add(10), "yellow")
plot.highlightNote("MM-3RD", root.add(11), "yellow")
plot.highlightNote("11/8", root.add(14), "cyan")
plot.highlightNote("5TH", root.add(18), "lightgreen")
plot.plotIsomorphic(start_note = 0, x_increment = -2, y_increment = 3)
```
