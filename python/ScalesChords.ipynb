## 12 EDO Scales and Chords

### Heptatonic Scales




```python
import tuning as tuning
import tuningplot as tuningplot
%matplotlib inline

tuning_12edo = tuning.createEqualDivisionTuning("12edo", 12, 1, ['0 C','1 C#','2 D','3 D#','4 E','5 F','6 F#','7 G','8 G#','9 A','10 A#','11 B'])
mode_major = tuning.Mode("Major", tuning_12edo, [0, 2, 4, 5, 7, 9, 11])
chord_triad = tuning.Chord(mode_major, [0, 2, 4])

plot = tuningplot.TuningPolarPlot(8, 2, 1)
plot.plotRatios([(1, 1), (3, 2), (4, 3), (5, 3), (5, 4), (6, 5)])
plot.plotTuning(tuning_12edo)
plot.plotMode(mode_major, 0, 2, "C Major Mode", True)
plot.plotChord(chord_triad, 0, "C Major Chord")
plot.plot()
plot.plotTuning(tuning_12edo)
plot.plotMode(mode_major, 0, 2, "C Major Mode", True)
plot.plotChord(chord_triad, 1, "D Minor Chord")
plot.plot()
```

```python


tuning_12edo = tuning.createEqualDivisionTuning("12edo", 12, 1, ['0 C','1 C#','2 D','3 D#','4 E','5 F','6 F#','7 G','8 G#','9 A','10 A#','11 B'])
mode_major = tuning.Mode("Major", tuning_12edo, [0, 2, 4, 5, 7, 9, 11])
chord_triad = tuning.Chord(mode_major, [0, 2, 4])

plot = tuningplot.TuningPlot('h', 16, 2)
plot.plotRatios([(1, 1), (2, 1)])
plot.plotTuning(tuning_12edo, 2)
plot.plotMode(mode_major, 0, 2, "C Major Mode", True)
plot.plotChord(chord_triad, 0, "C Major Chord")
plot.plotChord(chord_triad, 1, "D Minor Chord")
plot.plotChord(chord_triad, 2, "E Minor Chord")
plot.plotChord(chord_triad, 3, "F Major Chord")
plot.plotChord(chord_triad, 4, "G Major Chord")
plot.plotChord(chord_triad, 5, "A Minor Chord")
plot.plotChord(chord_triad, 6, "B Diminished Chord")
plot.plot(aspect=0.05)
```

```python
tuning_12edo = tuning.createEqualDivisionTuning("12edo", 12, 1, ['0 C','1 C#','2 D','3 D#','4 E','5 F','6 F#','7 G','8 G#','9 A','10 A#','11 B'])

plot = tuningplot.TuningPlot('h', 16, 2)
plot.plotRatios([(1, 1), (2, 1)])
plot.plotTuning(tuning_12edo, 2)

# Heptatonic First (2-2-1-2-2-2-1)
mode_ionian = tuning.Mode("Ionian (Major)", tuning_12edo, [0, 2, 4, 5, 7, 9, 11])
mode_dorian = tuning.Mode("Dorian", tuning_12edo, [0, 2, 3, 5, 7, 9, 10])
mode_phrygian = tuning.Mode("Phrygian", tuning_12edo, [0, 1, 3, 5, 7, 8, 10])
mode_lydian = tuning.Mode("Lydian", tuning_12edo, [0, 2, 4, 6, 7, 9, 11])
mode_mixolydian = tuning.Mode("Mixolydian", tuning_12edo, [0, 2, 4, 5, 7, 9, 10])
mode_aeolian = tuning.Mode("Aeolian", tuning_12edo, [0, 2, 3, 5, 7, 8, 10])
mode_locrian = tuning.Mode("Locrian", tuning_12edo, [0, 1, 3, 5, 6, 8, 10])
plot.plotMode(mode_ionian, 0, 2, "C Ionian (Major) Mode", True)
plot.plotMode(mode_dorian, 2, 2, "D Dorian Mode", True)
plot.plotMode(mode_phrygian, 4, 2, "E Phrygian Mode", True)
plot.plotMode(mode_lydian, 5, 2, "F Lydian Mode", True)
plot.plotMode(mode_mixolydian, 7, 2, "G Mixolydian Mode", True)
plot.plotMode(mode_aeolian, 9, 2, "A Aeolian Mode", True)
plot.plotMode(mode_locrian, 11, 2, "B Locrian Mode", True)

# Heptatonic Second (2-1-2-2-2-2-1)
mode_melodic_minor = tuning.Mode("Melodic Minor", tuning_12edo, [0, 2, 3, 5, 7, 9, 11])
plot.plotMode(mode_melodic_minor, 0, 2, "C Melodic Minor", True)

plot.plot(aspect=0.05)
```
